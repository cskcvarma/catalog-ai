```python
from huggingface_hub import HfApi, DatasetFilter, datasets
import os
from dotenv import load_dotenv

load_dotenv()

```

```python
HF_TOKEN = os.environ["HF_TOKEN"]
HF_ENDPOINT = os.environ["HF_ENDPOINT"]

hf_api = HfApi(
    endpoint=HF_ENDPOINT,
    token=HF_TOKEN
)
```

```python
data_filter: DatasetFilter = DatasetFilter(
    task_categories="text-classification",
)
datasets = hf_api.list_datasets(filter=data_filter)
```

```python
dataset_list = list(datasets)
```

```python
for dataset in dataset_list:
    print(dataset.cardData)
    if dataset.card_data is not None:
        print(dataset.card_data, dataset.id)
```
